# Imports

```python
import pandas as pd
import textstat
import matplotlib.pyplot as plt
```

# Understandability

## Rating

```python
# Load your CSV file
input_file = "empathy_prompts_with_responses.csv"
df = pd.read_csv(input_file)
print(df)
# Compute SMOG index and Flesch-Kincaid grade for each output
df["smog_index"] = df["Model Response"].apply(textstat.smog_index)
df["fk_grade"] = df["Model Response"].apply(textstat.flesch_kincaid_grade)

# Save updated DataFrame to a new CSV
output_file = "empathy_prompts_with_readability.csv"
df.to_csv(output_file, index=False)

print(f"✅ Readability scores added and saved to: {output_file}")
```

## Visualization

```python
# Load the CSV file
df = pd.read_csv("empathy_prompts_with_readability.csv")

# Create the plot
plt.figure(figsize=(12, 7))

# Overall title
plt.suptitle(
    "Readability Scores for Doctor–Patient Outputs Generated by ChatGPT-4o",
    fontsize=14, fontweight='bold', y=1.08
)

# --- Plot 1: SMOG Index ---
plt.subplot(2, 2, 1)
plt.hist(df["smog_index"], bins=15, edgecolor="black")
#plt.axvline(x=10, color="red", linestyle="--", label="Recommended Max (10)")
plt.title("SMOG Index Distribution")
plt.xlabel("SMOG Index")
plt.ylabel("Number of Responses")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

# --- Plot 2: Flesch-Kincaid Grade ---
plt.subplot(2, 2, 2)
plt.hist(df["fk_grade"], bins=15, edgecolor="black", color="skyblue")
#plt.axvline(x=8, color="red", linestyle="--", label="Recommended Max (8)")
plt.title("Flesch-Kincaid Grade Distribution")
plt.xlabel("FK Grade Level")
plt.ylabel("Number of Responses")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

# --- SMOG Interpretation Text ---
smog_text = (
    "  SMOG Index Interpretation:\n"
    "  • 6–8   → Middle School Level\n"
    "  • 9–12  → High School Level\n"
    "  • 13–16 → College/Undergraduate Level\n"
    "  • >16   → Graduate/Advanced Academic Level"
)
plt.subplot(2, 2, 3)
plt.axis("off")
plt.text(0, 0.9, smog_text, fontsize=10, verticalalignment='top', family='monospace')

# --- FK Grade Interpretation Text ---
fk_text = (
    "  Flesch-Kincaid Grade Interpretation:\n"
    "  • 6–8   → Middle School Level\n"
    "  • 9–12  → High School Level\n"
    "  • 13–16 → College/Undergraduate Level\n"
    "  • >16   → Graduate/Advanced Academic Level"
)
plt.subplot(2, 2, 4)
plt.axis("off")
plt.text(0, 0.9, fk_text, fontsize=10, verticalalignment='top', family='monospace')


# Layout
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```
